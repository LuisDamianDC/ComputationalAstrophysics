```python
# -*- coding: utf-8 -*-
"""
Created on Wed Apr  3 14:47:26 2024

@author: LuisDami√°nDC_
"""
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
N=100
L=10.0
Nc=0.8
u=np.zeros(N+2)
print(len(u))
dx=L/N
a=0.6 #velocidad
```

```python
tf=0.4 #tiempo final

x=np.arange(0, L+2*dx, dx)
up=np.zeros(len(x))

def u0i(x):
    return np.exp(-(x-2))

def u0a(x):
    return np.exp(-(x-2)**2)

def primtocons(p):
    return p

def primtoflux(p):
    return p**2

def constoprim(u):
    return u
```

```python
u=u0a(x)
ui=u0a(x)
t=0
U=primtocons(u)
while(t<tf):
    U[0]=U[N]
    U[N+1]=U[1]
    dt=Nc*dx/(2*max(U)) #cambiarlo por el maximo del vector u
    t=t+dt
    P=constoprim(U)
    F=primtoflux(P)
    for i in range(1, N+1):
        Fr=(F[i]+F[i+1])/2-(dx/(2*dt))*(U[i+1]-U[i])
        Fl=(F[i-1]+F[i])/2-(dx/(2*dt))*(U[i]-U[i-1])
        up[i]=U[i]-(dt/dx)*(Fr-Fl)
    U=up
```

```python
plt.plot(x, ui, 'y')
plt.plot(x, U, 'r')
plt.plot(x, P, 'g')
plt.plot(x, F, 'b',)
```
